Funzione per ottenere la Definizione dal codice

```python
def get_definizione(code):
    return df_codes[df_codes["NOME_CAMPO"]==code]["DEFINIZIONE"].values[0]
```

Funzione per salvare tuttele definizioni in una lista

```python
col_def_list = []
for col_name in df.columns:
    try:
        def_name = get_definizione(col_name)
        col_def_list.append(def_name)
        print def_name
    except:
        print "not found"
```

Re-define the column names in the data-frame 

```python
df.columns = col_def_list
```
